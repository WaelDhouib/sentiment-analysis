### Generating Graph:

```python
import networkx as nx
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('..\data\processed\network.csv')

df.describe()



df.userFromId=df.userFromId.apply(lambda x: int(x))
df.userToId=df.userToId.apply(lambda x: int(x))


G = nx.Graph()

G.add_nodes_from(df['userFromId'])


temp = zip(df['userFromId'],df['userToId'])
G.add_edges_from(temp)

dfLookup = df[['userFromName','userFromId']].drop_duplicates()

dfLookup.head()
for userId in dfLookup['userFromId']:
    temp = dfLookup['userFromName'][df['userFromId']==userId]
    G.node[userId]['userName'] = temp.values[0]

nx.draw(G, pos=nx.spring_layout(G,k=.90),node_color='k',edge_color='g')


pos=nx.spring_layout(G,k=.12)

centralScore = nx.betweenness_centrality(G)
```

```python
plt.hist(list(dict(nx.degree(G)).values()))
plt.xlabel('Degrees')
plt.ylabel('Number of Subjects')
plt.show()
dict(nx.degree(G)).values()
```

### Sentiment Analysis:

```python
from textblob import TextBlob
import nltk
t=''
for i in range(len(text)):
    t += text[i]
```

```python
nltk.download('punkt')
blob = TextBlob(t)
blob.tags
blob.noun_phrases
s=0
for sentence in blob.sentences:
    s+=sentence.sentiment.polarity
print("Polarity overall: ",s)
```

```python
import numpy as np
sentiment = np.zeros(5)
for sentence in blob.sentences:
    if sentence.sentiment.polarity>0.75:
        sentiment[0]+=1
    if sentence.sentiment.polarity>0:
        sentiment[1]+=1
    if sentence.sentiment.polarity==0:
        sentiment[2]+=1
    if sentence.sentiment.polarity<0:
        sentiment[3]+=1
    if sentence.sentiment.polarity<-0.75:
        sentiment[4]+=1
```

```python
import matplotlib.pyplot as plt
 
#Data to plot
labels = 'Very positive', 'Positive', 'Neutral', 'Negative', 'Very negative'
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red']
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice
 
#Plotting a pie chart
plt.pie(sentiment, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
 
plt.axis('equal')
plt.show()
```
